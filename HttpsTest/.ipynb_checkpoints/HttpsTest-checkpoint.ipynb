```python
PROXY_POOL_URL = 'http://localhost:5555/random'

def get_proxy():
    try:
        response = requests.get(PROXY_POOL_URL)
        if response.status_code == 200:
            return response.text
    except ConnectionError:
        return None
```

```python
def getHttps3(url):
    print(url)
    #time.sleep(numpy.random.rand()*5)
    
    try:
        ip=get_proxy()
        print(ip)
        proxies = {"http": "http://"+ip, "https": "http://"+ip}
        response = requests.get(url, mHeader.random(), proxies=proxies, timeout=20) 
        response.raise_for_status()  # 如果响应状态码不是 200，就主动抛出异常
        #print(response.text)
    except requests.RequestException as e:
        print(e)
    else:
        for i in response.json()['data']:
            print(i['title'])
            print(i['rate'])
            print(i['url'])
    
    
```
